```python
import os
import matplotlib.pyplot as plt


audio_files_directory = r"C:\Users\ishas\OneDrive\Documents\ML Project\Data\genres_original"


def count_files_by_genre(directory):
    genres = {}
    for genre in os.listdir(directory):
        if os.path.isdir(os.path.join(directory, genre)):
            genres[genre] = len(os.listdir(os.path.join(directory, genre)))
    return genres


audio_files = count_files_by_genre(audio_files_directory)
genres = list(audio_files.keys())
file_counts = list(audio_files.values())


plt.figure(figsize=(8, 6))
plt.pie(file_counts, labels=genres, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Audio Files by Genre')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
